## Q2. Indexing and searching

Index with minsearch (`content` = text field, `filename` = keyword field).

Query: *How does the agentic loop keep calling the model until it stops?*

What's the `filename` of the first result?

```python
from minsearch import Index

index = Index(
    text_fields=["content"],
    keyword_fields=["filename"],
)
index.fit(documents)

query_q2 = "How does the agentic loop keep calling the model until it stops?"
results_q2 = index.search(query_q2, num_results=5)
results_q2[0]["filename"]
```
